# Introdução

Esse notebook faz uma análise exploratória (EDA) da base de dados [2020 Kaggle Machine Learning & Data Science Survey](https://www.kaggle.com/c/kaggle-survey-2020/data).

Essa base contém 355 atributos relativos a respostas de um survey de 35 perguntas relativas a Machine Learning e Data Science. Essas perguntas foram respondidas por 20.037 pessoas e são organizadas de acordo com a estrutura da tabela abaixo.

| Questões | Descrição                                                          |
|----------|--------------------------------------------------------------------|
| 1-6      | Perguntas gerais sobre o usuário (idade, gênero,país,...)          |
| 7-13     | Linguagens de programação, IDE e hardware utilizados               |
| 14-19    | Frameworks, bibliotecas e abordagens (NLP, detecção de objetos...) |
| 20-25    | Informações da empresa em que trabalha                             |
| 26-32    | Ferramentas de cloud, big data e business inteligence              |
| 33-34    | AutoML                                                             |
| 35-36    | Ferramentas para acompanhamento de experimentos e deploy           |
| 37       | Plataformas para aprendizado da área                               |
| 38       | Ferramentas de análise de dados                                    |
| 39       | Redes sociais para acompanhamento da área                          |

De acordo com essas questões, os atributos da base são organizados utilizando o número das questões (e.g. Q1, Q2, Q3). Para questões em que o usuário pode escolher mais de uma resposta, cada atributo representa uma (e.g. Q7 - "quais linguagens de programação você utiliza?" divide-se nos atributos Q7_Part_1, Q7_Part_2, ..., em que cada parte é uma resposta). Além disso, o atributo "Time from Start to Finish (seconds)" mostra o tempo total que o usuário levou para completar o survey.
